```python
import json
import pymongo
from utils import build_key
import wikipedia

wikipedia.set_lang("fr")

client = pymongo.MongoClient()

FILEPATH_WKPD = "streets/wkpd/out/streets_from_wkpd.json"
```

```python
collection_voie = client["visitp"]["streets_voies"]
```

```python
with open(FILEPATH_WKPD) as f:
    data_wkpd = json.load(f)

len(data_wkpd)
```

```python
wkpd_keys = set(data_wkpd.keys())
```

```python
found = []
not_found = dict()
for doc in collection_voie.find():
    l_longmin = doc["fields"]["l_longmin"]
    key = build_key(l_longmin)
    if key in wkpd_keys: found.append(l_longmin)
    else: not_found[key] = l_longmin

print("    found : ", len(found))
print("not found : ", len(not_found))

    
```

```python
collection_voie.count()
```

```python
for key, street in not_found.items():
    print(street)
    try:
        res = wikipedia.page(street)
        print("\t ! :", res.url)
    except:
        try:
            res = wikipedia.search(street)
            print("\t ? :", res)
        except:
            print("\t XXX")
```
